```python
import pandas as pd
import numpy as np
import ast # to make the string representation of list back to list
df = pd.read_csv("preprocessed_data.csv")
df['text'] = df['text'].apply(ast.literal_eval)
```

```python
df['next_word'] = df['next_word'].apply(lambda x: [x])

df.head()
```

```python
df.head()
```

```python

max_length = df['text'].str.len().max()
index = df['text'].str.len().idxmax()
print("The longest text is:", df['text'].iloc[index])
print("The length of the longest text is:", len(df['text'].iloc[index]))
print(max_length)
```

```python
df.info()
```

total words 

```python
all_words = set()
for text in df['text']:
    all_words.update(text)
for text in df['next_word']:
    all_words.update(text)
all_words = list(all_words)
print(len(all_words))
```

```python
all_words
```

```python
all_words.index('charismatic')
```

```python
id = {word: idx+1 for idx, word in enumerate(all_words)}  # idx +1 because we want to reserve 0 for padding
print(id['charismatic'],id['impose'])
```

```python
word_id = {idx: word for word, idx in id.items()}
print(word_id[1], word_id[22])
```

```python
x  = pd.Series(df['text'].apply(lambda text: [id[word] for word in text]))
y = pd.Series(df['next_word'].apply(lambda text: [id[word] for word in text]))
print(x[45])
print(y[45])
```

```python
print(x.shape)
print(y.shape)
```

```python
print(x.str.len().max())
print(y.str.len().max())
```

```python
df=pd.DataFrame({'text': x, 'next_word': y})
df.head()
```

```python
max_len = df["text"].str.len().max()
df['text'] = df['text'].apply(lambda x: [0]*(max_len - len(x))+x)  # padding with 0
print(df['text'][45])
```

```python
df.info()
```

```python
df.to_csv("final_data.csv", index=False)
```
