```python
import pandas as pd
```

```python
with open("Next Word Predictor Dataset.txt", "r", encoding="utf-8") as f:
    data = f.readlines()
```

```python
df = pd.DataFrame(data, columns=["text"])
```

```python
df.head(10)
```

```python
df['text'] = df['text'].str.strip()
```

```python
df.shape
```

```python
df.isnull().sum()   
```

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True) 
```

```python
data = []
for i in range(0, df.shape[0]):
    df['text'] = df['text'].str.lower()
    words_list = df['text'].iloc[i].split() 
    for j in range(1, len(words_list) + 1):
        data.append(words_list[:j])

```

```python
new_df = pd.DataFrame({"text": data})
```

```python
new_df.head(30)
```

```python
new_df.shape
```

```python
new_df.duplicated().sum()
```

```python
new_df.drop(new_df[new_df['text'].str.len() == 1].index, inplace=True)
```

```python
new_df.drop_duplicates(inplace=True)
```

```python
new_df['next_word'] = new_df['text'].str[-1]
```

```python
new_df['text'] = new_df['text'].str[:-1]
```
